文字列`'...'`や`"..."`のはじめの`'`（クォーテーション）または`"`（ダブルクォーテーション）の前に文字`f`をつけた文字列を**f-string（f-文字列）**という。f-stringは次のような形をとる。f-stringにある`{}`（半角中括弧）内には変数や数式を入れることができる：

```
f'...{変数または数式}...{変数または数式}...'
```

上のf-stringは以下と同じである：

```
'...' + str(変数または数式) + '...' + str(変数または数式) + '...'
```

変数や数式が混在する文字列を作成する場合，上のf-stringを用いた形式の方が下の`+`で繋ぐ形式よりはるかに簡単である。実際，現在ではf-stringを用いる形式が標準となっている。しかしこれだけではわかりにくいので，例を試してみよう。
